# Groupby

#### Groupby groups data into smaller sections

```python
df.groupby('sex').mean()
```

```python
df.groupby('sex').describe()
```

```python
df.groupby('sex').describe()['tip']
```

```python
df.groupby(['sex','day'])
```

```python
df.groupby(['sex','day']).mean()
```

```python
g = df.groupby('sex')
g
```

#### So as you can see, g is a groupby abject

```python
g.groups.keys()
```

```python
g.get_group('Male')
```
