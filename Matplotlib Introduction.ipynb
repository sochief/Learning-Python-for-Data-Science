```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(0, np.pi*4, num = 200)
```

```python
x
```

#### Here we created 200 numbers between 0 and Pi*4

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

##### if you dont want to see the first output u type

```python
plt.plot(x,y)
plt.show()
```

#### Also we can give labels for each of the parameters, f.e:

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

#### For example, you have lots of functions, so it would be good, if you will have a legend
##### While creating the graph, you should give a legemd for that as well

```python
plt.plot(x,y, label = 'sin(x)')
plt.xlabel('X')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

#### okay, so and if you want to, for example, change the color of the line you type
#### b	blue                                                                     g	green
#### r	red                                                                        c	cyan
#### m	magenta                                                             y	yellow
#### k	black                                                                    w	white

```python
plt.plot(x,y,'c', label = 'sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

###### So let's add one more function, for example z, which will be equal to cos(x)

```python
z = np.cos(x)

plt.plot(x,y,'c', label = 'sin(x)')
plt.plot(x,z,'m', label = 'cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

###### You can change x limmits and y limits as well, so let's do that

```python
z = np.cos(x)
plt.plot(x,y,'c',label='sin(x)')
plt.plot(x,z,'m',label='cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,6]) # so basically, we make out a limit here
plt.legend()
plt.show()
```

#### If we are creating a sublot with paarameters 1,2,1 we say, that we are going to create a subplot with 1 row and 2 coloumns and we are going to plot on the first subplot 

```python
plt.subplot(1,2,1)
plt.plot(x,y,'c')

plt.subplot(1,2,2)
plt.plot(x,z,'b')

plt.tight_layout() #will make graph look nicer
plt.show()
```

###### or we can make a graph a bit larger

```python
plt.figure(figsize=(10,5))# first parameter is the width and the second is the height

plt.subplot(1,2,1)
plt.plot(x,y,'c')

plt.subplot(1,2,2)
plt.plot(x,z,'b')

plt.tight_layout() #will make graph look nicer
plt.show()  
```

##### We can use Objective oriented format, so first we create an object, and then we work with it

```python
fig = plt.figure()

# figure is the whole graph
axes = fig.add_axes([0.1,0.1,0.9,0.8])
```

#### we have created our empty axes so now will add things to it

```python
fig = plt.figure()

# figure is the whole graph
axes = fig.add_axes([0.1,0.1,0.9,0.8])
axes.plot(x,y,label ='sin(x)')
axes.legend()
axes.set_xlabel('x')
axes.set_ylabel('sin(x)')
axes.set_title('My first graph')

plt.show()
```
